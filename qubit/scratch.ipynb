```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot1():
    x = np.linspace(-3.5, 3.5, 400)
    V = 0.5 * np.power(x, 2)

    n_levels = 6
    energies = [n + 0.5 for n in range(n_levels)]

    max_energy = energies[-1]
    label_x = np.sqrt(2 * max_energy) + 0.2

    fig, ax = plt.subplots(figsize=(3, 4), dpi=300)
    fig.patch.set_facecolor("none")
    ax.patch.set_facecolor("none")

    ax.plot(x, V, "k-", linewidth=2)

    for n, E in enumerate(energies):
        x_left = -np.sqrt(2 * E)
        x_right = np.sqrt(2 * E)

        ax.hlines(
            E, xmin=x_left, xmax=x_right, colors="k", linestyles="--", linewidth=2
        )

        # ax.text(
        #     label_x, E, f"  $|{n}\\rangle$", verticalalignment="center", fontsize=18
        # )

        # if n == 0:
        #     ax.text(
        #         label_x, E, f"$|g\\rangle$", verticalalignment="center", fontsize=18
        #     )

    ax.set(xticks=[], yticks=[])
    ax.spines[["left", "right", "top", "bottom"]].set_visible(False)
    ax.grid(False)

    plt.tight_layout()
    plt.savefig("./fig1.svg")
    plt.show()


plot1()
```
